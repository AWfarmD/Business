```python
import pandas as pd
```

```python
df = pd.read_json('Cell_Phones_and_Accessories_5.json', lines=True)
```

```python
df.head()
```

```python
df['reviewTime'] = df['reviewTime'].apply(pd.to_datetime)
```

```python
df.head()
```

```python
df_2018 = df[df.reviewTime > '2018-01-01']
```

```python
df_2018.head()
```

```python
df_2018.to_csv('Cell and Accessories.csv')
```

* Getting 2017 data

```python
df_2017 = df[(df.reviewTime >= '2017-06-01') &
             (df.reviewTime <= '2017-12-31')]
```

```python
df_2017.reset_index(drop=True, inplace=True)
```
